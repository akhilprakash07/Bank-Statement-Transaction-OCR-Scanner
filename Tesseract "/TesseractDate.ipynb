```python
from pdf2image import convert_from_path
import pytesseract
import re
import pandas as pd

pdf_path = '/content/SBI.pdf'

images = convert_from_path(pdf_path)

texts = [pytesseract.image_to_string(image) for image in images]

full_text = "\n".join(texts)

pattern = r'(?:"(\d{1,2} \w{3}\s*\d{4})"|\b(\d{1,2} \w{3} \d{4})\b)'

matches = re.findall(pattern, full_text) # Use full_text instead of text

txn_dates = [date for match in matches for date in match if date]

print("Txn Dates:", txn_dates)

max_len = len(txn_dates)

data = {
    'Txn Dates': txn_dates
}

df = pd.DataFrame(data)

df.to_csv('/content/extracted_data.csv', index=False)
```
